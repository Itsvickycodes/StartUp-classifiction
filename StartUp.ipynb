```python
#check missing value for categorical columns in dataset
CATEGORICAL = list(filter(lambda x: data[x].dtype=='object',data.columns))
def show_stats(columns):
    stat = {}
    for col in columns:
        stat[col] =  [data[col].nunique(), data[col].isna().mean()*100, data[col].dtype]
    return pd.DataFrame.from_dict(stat, orient='index', columns=['Unique',  'Missed values in %', 'DType'])
show_stats(CATEGORICAL)
```

```python
def str_to_int(s: str):
    try:
        s = s.replace(',', '')
        if '-' in s:
            s = 0
        return int(s)
    except ValueError:
        return s
    except AttributeError:
        return s
data['funding_total_usd'] = data['funding_total_usd'].apply(str_to_int)
```

```python
def correlation_fun(ds,threshold):
    corr_col=set()
    corr_mat=ds.corr()
    for i in range(len(corr_mat.columns)):
        for j in range(i):
            if abs(corr_mat.iloc[i,j])>threshold:
                colname=corr_mat.columns[i]
                corr_col.add(colname)
    return corr_col
```

```python
threshold=0.5
correlation_fun(data.drop("status",axis=1),threshold)
```

# train and test the model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape)
```

# KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
